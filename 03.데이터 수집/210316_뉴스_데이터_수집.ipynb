## 데이터 수집 뉴스1

```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent':'Mozilla/5.0'}
url = 'https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001'

html = requests.get(url,headers=headers).text

soup = BeautifulSoup(html, 'html.parser')
type06 = soup.find('ul',{'class':'type06_headline'})
dl = type06.find_all('dl')
for item2 in dl:
    try:
        img = item2.find('dt',{'class':'photo'}).find('img')
        print('img=',img['src'])
        print('--------------------------')
    except:
        print('No image')
```

```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent':'Mozilla/5.0'} #header정보 보는 법 F12 개발자 화면에서 네트워크를 누르고 url 엔터 user-agent 정보에서 확인
url = 'https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001'
html = requests.get(url,headers=headers).text #header를 지정해줘서 정상적으로 브라우저에서 접근하는 것처럼 처리
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup #html 구조가 보기 좋게 정리. 브라우저 화면에 보이는 것처럼.
```

```python
type06 = soup.find('ul',{'class':'type06_headline'})
type06 #ul태그에 있는 데이터 수집
#페이지가 어떻게 구성되었는지 확인 > 검사페이지에서 데이터 영역 확인
```

```python
dl = type06.find_all('dl') #ul 아래에 있는 모든 dl태그를 불러옴
dl
```

## 데이터 수집 뉴스2

```python
dl = type06.find_all('dl')
for item2 in dl:
    link = item2.find('dt',{'class':''}).find('a')
    print(link['href'])
    print(link.text.replace('\t','').replace('\n','').replace('\r','')[1:len(link.text)+1]) #[0]은 공백
    #.strip()도 사용 가능
```
