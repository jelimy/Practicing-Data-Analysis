## 실습) 10개종목 시세데이터

1. 종목 10개, page 5개
2. 시세 DataFrame 만들기: 종목코드, 날짜, 종가, 전일대비, 시가, 고가, 저가, 거래량

https://finance.naver.com/item/sise_day.nhn?code=005930&page=1

```python
import pandas as pd
import requests
import time
import urllib.request
```

```python
url = 'https://finance.naver.com/item/sise_day.nhn?code=005930&page=1'
b_url = 'https://finance.naver.com/item/sise_day.nhn?code={}&page={}'
headers = {"user-agent":"Mozilla/5.0"}
code = ['005930','051910','214390','185750','006400','063160','068290','096770','006980','035720']
time.sleep(0.5)

#resp = b_url에 for문을 돌려서 필요한 모든 url을 추출해 불러옴
#df = 불러온 html 데이터의 텍스트를 판다스를 이용해 데이터 프레임 형태로 읽어옴
#df를 다시 정의, 각 데이터의 null값을 제거
#fin에 이러한 과정을 모두 넣는다.

fin = []

for n in range(len(code)):
    for i in range(5):
        resp = requests.get((b_url.format(code[n],i+1)), headers=headers)
        df = pd.read_html(resp.text)
        df = df[0].dropna()
        fin.append(df)
```

```python
fin
```

```python
#fin의 형태 : [(종목코드1x(1-5page))x10개 종목] 형태
```

```python
#각 종목 코드별로 리스트를 묶어준다.
n=5
code_list = []
for row in range((len(fin) + n - 1)//n):
    data=fin[row*n:(row +1)*n]
    code_list.append(data)
```

```python
#각 종목 코드별로 리스트 묶은 것을 각각 데이터 프레임으로 병합
result = []

for v in range(len(code_list)):
    a = pd.concat(code_list[v],ignore_index=True) #종목코드별로 인덱스값 새로 설정
    result.append(a)
```

```python
result[0]
```

```python
result2 = pd.concat(result, keys = code, names = ['종목코드', ''])
result2
```
