## 실습
1) youtube.com \
2) 화면에서 키워드 입력 받아서 검색 결과
- input()
- url(예) > https://www.youtube.com/results?search_query=파이썬+데이터분석 

3) DataFrame으로 만들기
- 개별 영상 타입 : \
    채널명, 영상제목, 영상url, 조회수, 게시일
- 재생목록보기 타입 : \
    채널명, 제목, 재생목록 url, 구성영상수

```python
# 개별 영상 타입
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver

driver = webdriver.Chrome()

base_url = 'https://www.youtube.com/results?search_query='
url = base_url + input('')

driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

text_wrapper = soup.find_all('div',{'class':'text-wrapper style-scope ytd-video-renderer'})

video = []

for link in text_wrapper:
    channel = link.find('div',{'id':'tooltip'}).text.strip()
    th = link.find('a',{'id':'video-title'})
    title = th.get('title')
    href = 'https://www.youtube.com' + th.get('href')
    click = link.find_all('span',{'class':'style-scope ytd-video-meta-block'})[0].text
    date = link.find_all('span',{'class':'style-scope ytd-video-meta-block'})[1].text
    
    video.append([channel, title, href, click, date])
    
df_video = pd.DataFrame(video, columns=['채널명', '영상제목', '영상url', '조회수', '게시일'])
df_video.set_index('채널명', inplace=True)
df_video
```

```python
# 재생목록보기 타입
```

```python
item_section = soup.find_all('ytd-playlist-renderer',{'class':'style-scope ytd-item-section-renderer'})

playlist = []

for link in item_section:
    channel = link.find('div',{'id':'tooltip'}).text.strip()
    title = link.find('span',{'id':'video-title'}).get('title')
    href = 'https://www.youtube.com' + link.find('a',{'class':'yt-simple-endpoint style-scope yt-formatted-string'}).get('href')
    count = link.find('yt-formatted-string',{'class':'style-scope ytd-thumbnail-overlay-side-panel-renderer'}).text
    
    playlist.append([channel, title, href, count])
    
df_playlist = pd.DataFrame(playlist, columns=['채널명', '재생목록 제목', '재생목록 url', '구성영상수'])
df_playlist.set_index('채널명', inplace=True)
df_playlist
```
