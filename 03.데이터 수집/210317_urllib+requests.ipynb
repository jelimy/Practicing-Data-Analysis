## urllib.request :URL을 열기 위한 확장 가능한 라이브러리

http://docs.python.org/ko/3.10/library/urllib.request.html

```python
import urllib.request
with urllib.request.urlopen('http://www.python.org/') as f:
    print(f.read(300))
```

```python
import urllib.request
with urllib.request.urlopen('http://www.python.org/') as f:
    print(f.read(100).decode('utf-8'))
```

```python
import urllib.request
import urllib.parse
params = urllib.parse.urlencode({'mode':'LSD', 'mid':'sec','sid1':'001', 'date':'20210317', 'page':'3'})
params
```

```python
url = 'https://news.naver.com/main/list.nhn?%s'% params

opener = urllib.request.build_opener()
opener.addheaders = [('User-agent','Mozilla/5.0')]
res = opener.open(url)

res
```
