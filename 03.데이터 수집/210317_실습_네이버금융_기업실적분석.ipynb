```python
import requests
from bs4 import BeautifulSoup
```

```python
#사이트 불러오기
headers = {'User-Agent':'Mozilla/5.0'}
url = 'https://finance.naver.com/item/main.nhn?code=005930'

#필요한 데이터 위치
html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html,'html.parser')
tb_type1 = soup.find('table',{'class':'tb_type1 tb_num tb_type1_ifrs'})
tbody = tb_type1.find('tbody')
td = tbody.find_all('td')
```

```python
#필요한 데이터 수치만 불러오기
issue = []

for i in range(0,len(td)):
    try:
        val = td[i].text.replace('\t','').replace('\n','')
        print(val)
        print('--------')
        issue.append(val)
    except:
        print('None')
```

```python
issue
```

```python
#데이터프레임 인덱스로 사용할 데이터 불러오기
th = tbody.find_all('th')

row_name = []

for i in range(0,len(th)):
    try:
        h = th[i].text.replace('\t','').replace('\n','')
        print(h)
        print('--------')
        row_name.append(h)
    except:
        print('None')
```

```python
row_name
```

```python
#데이터프레임 컬럼명으로 사용할 데이터 불러오기
html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html,'html.parser')
tb_type1 = soup.find('table',{'class':'tb_type1 tb_num tb_type1_ifrs'})
thead = tb_type1.find('thead')
tr = thead.find_all('tr')
date = tr[1].find_all('th')
```

```python
col_name = []

for i in range(0,len(date)):
    try:
        val2 = date[i].text.replace('\t','').replace('\n','')
        #val2 = date[i].test.strip()
        print(val2)
        print('--------')
        col_name.append(val2)
    except:
        print('None')
```
