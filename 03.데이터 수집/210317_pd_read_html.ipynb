## pandas로 html 테이블 형태 바로 읽어오는 방법
- pd.read_html()

```python
import pandas as pd

url = 'https://finance.naver.com/item/main.nhn?code=005930'

df = pd.read_html(url, encoding='cp949') # 한글을 읽기위한 것. utf-8로 읽을 수 있는 한글도 있다.
df[3]
#테이블 형태로 되어 있을 때 유용
```

```python
col_list = []
for col in df[3].columns:
    col_list.append(col[1])

df[3].columns
```

```python
df[3].columns=col_list
df[3]
```
