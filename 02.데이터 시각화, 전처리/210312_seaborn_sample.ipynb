## lmplot

```python
import seaborn as sns
sns.set_theme(style='ticks')

#데이터 로드
df=sns.load_dataset('anscombe')
df
```

```python
type(df)
```

```python
sns.lmplot(x='x',y='y', data=df)
```

```python
sns.lmplot(x='x',y='y', data=df,
          col='dataset', hue='dataset') #col=dataset에 따라 분류된 값 / hue=dataset에 따라 색을 다르게 함
```

```python
sns.lmplot(x='x',y='y', data=df, hue='dataset')
```

```python
sns.lmplot(x='x',y='y', data=df,
          col='dataset', hue='dataset', col_wrap=3) #col_wrap 배치 순서에 대한 옵션
```

```python
sns.lmplot(x='x',y='y', data=df,
          col='dataset', hue='dataset',
          col_wrap=2, ci=None, palette='muted', height=3,
          scatter_kws={'s':50, 'alpha':0.1})
#ci 뒤에 그림그려주는 것.
```

## Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='whitegrid')

diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
f, ax = plt.subplots(figsize=(6.5,6.5))
sns.despine(f, left=True, bottom=True)
```

```python
sns.scatterplot(x="carat", y="price", data=diamonds,
               hue='clarity')
```

```python
sns.scatterplot(x="carat", y="price", data=diamonds,
               hue='clarity', size='depth')
```

```python
sns.scatterplot(x="carat", y="price", data=diamonds,
               hue='clarity', size='depth',
               palette='ch:r=-.2, d=.3_r')
```

```python
sns.scatterplot(x="carat", y="price", data=diamonds,
               hue='clarity', size='depth',
               palette='ch:r=-.2, d=.3_r',
               hue_order=clarity_ranking,
               sizes=(1,8), ax=ax)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

diamonds = sns.load_dataset("diamonds")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```

## Lineplot, Timeseries plot

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

## Boxplot

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python
import seaborn as sns

sns.boxplot(x='day', y='total_bill',data=tips)
```

## heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()

flights_long = sns.load_dataset('flights')
flights_long
```

```python
flights=flights_long.pivot('month','year','passengers')
flights
```

```python
f, ax = plt.subplots(figsize=(9,6))
sns.heatmap(flights, annot=True, fmt='d', linewidths=.5)
```
