```python
#plot 함수 만들기
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict) ##*는 튜플형태라는 것을 지정, **sms 딕셔너리형태라는 것을 지정
    return out

#def my_plotter1(ax, data1, data2, param_dict):
#   return ax.plot(data1, data2, ** param_dict)
```

```python
#my_plotter 함수 실행
data1, data2, data3, data4 = np.random.randn(4,100)
fig,ax=plt.subplots(1,1)
my_plotter(ax, data1, data2, {'marker':'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {'marker':'x', 'color':'green'})
my_plotter(ax2, data3, data4, {'marker':'o', 'color':'red'})
#param_dict 부분에 시각화 옵션을 설정
```

```python
plt.hist(x, 50, density=True, facecolor='g', alpha=1) #알파값은 진하게 흐리게 like 투명도
plt.xlabel('Smarts')
plt.ylabel('Probatility')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100, \ \sigma=15$') #텍스트 문장 좌표
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
np.random.seed(19680801) #seed 랜덤 중 하나를 번호 지정해서 같은 값을 만들기 위함
```

```python
np.random.seed(19680801)

y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

#여러 도표의 간격 조절
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.35)

plt.show()
```
