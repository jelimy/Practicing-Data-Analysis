## german_sample2 만들기

```python
german_sample2 = german[['Credit_amount','Present_residence','Purpose']]
german_sample2.head(3)
```

```python
german_grouped = german_sample2['Credit_amount'].groupby(german_sample2['Present_residence'])
german_grouped
```

```python
german_grouped.mean() #평균값
```

```python
german_grouped.count()
```

```python
german_grouped2 = german_sample2['Credit_amount'].groupby([german_sample2['Purpose'], german_sample2['Present_residence']])
german_grouped2
```

```python
german_grouped2.mean()
```

```python
german_sample2.count()
```

```python
german_sample3 = german[['Present_residence','Personal_status___sex','Credit_amount']]
german_sample3
```

```python
german_sample3.groupby('Present_residence').mean()
```

```python
for type, group in german_sample3.groupby('Present_residence'):
    print('type=', type)
    print('group=', group.head(3))
```

```python
group_data=german_sample3.groupby('Present_residence')
for group in group_data:
    print('group=', group)
```

```python
group_data=german_sample3.groupby('Present_residence')
for type, group in group_data:
    print('type=', type)
    print('group=', group.head(3))
```

```python
group_data=german_sample3.groupby(['Present_residence','Personal_status___sex'])

for (residence, sex), group in group_data:
    print((residence,sex))
    print(group.head(3))
```
