```python
from pandas import Series, DataFrame
import pandas as pd

data={'국어':[76,90,64,70,98], '영어':[65,75,70,82,93], '수학':[80,98,62,84,85]}
data
```

```python
df=DataFrame(data); df
```

```python
len(df)
```

```python
df.columns
```

```python
df['총점']=''
df['평균']=''
df['등급']=''

for i in range(0,len(df)):
    df.iloc[i,3]=df.iloc[i,0]+df.iloc[i,1]+df.iloc[i,2]
    
    df.iloc[i,4]=df.iloc[i,3]/3
    
    if df.iloc[i,4] > 90:
        df.iloc[i,5] = 'S등급'
    elif df.iloc[i,4] > 80:
            df.iloc[i,5] = 'A등급'
    elif df.iloc[i,4] >70:
        df.iloc[i,5] = 'B등급'
    else :
        df.iloc[i,5] = 'C등급'
    
df
```
